```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file_US_videos= "Resources/USvideos.csv"
USvideos = pd.read_csv(csv_file_US_videos)
USvideos.columns

US_videos_cleaned=USvideos[['video_id', 'trending_date','title','category_id','channel_title','publish_time','views', 'likes', 'dislikes', 'comment_count','description']]
US_videos_cleaned.head()
```

```python
US_videos_df = US_videos_cleaned.drop(['video_id','description'], axis = 1)
US_videos_df.head()
```

```python
json_file = "Resources/US_category_id.json"
US_category_id = pd.read_json(json_file)

Item_id=US_category_id["items"]
new_Item_id=pd.json_normalize(Item_id)
new_Item_id.head()

us_category_df = new_Item_id[['id','snippet.title']]
us_category_df = us_category_df.rename({"id" : "category_id", "snippet.title": "category"},axis='columns')
us_category_df
```

```python
us_category_df['category_id'] = us_category_df['category_id'].astype(int)
```

```python
us_youtube_df = pd.merge(US_videos_df, us_category_df, on=['category_id'])
```

```python
summay_df= us_youtube_df[['title','views','likes','dislikes','comment_count','category']]
```
