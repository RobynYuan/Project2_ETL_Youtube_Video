```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file_US_videos= "Resources/USvideos.csv"
USvideos = pd.read_csv(csv_file_US_videos)
USvideos.columns

US_videos_cleaned=USvideos[['video_id', 'trending_date','title','category_id','channel_title','publish_time','views', 'likes', 'dislikes', 'comment_count','description']]
US_videos_cleaned = US_videos_cleaned.drop_duplicates(subset='video_id', keep='last')
US_videos_cleaned.head()

```

```python
json_file = "Resources/US_category_id.json"
US_category_id = pd.read_json(json_file)

Item_id=US_category_id["items"]
new_Item_id=pd.json_normalize(Item_id)
Categories=new_Item_id [['id', 'snippet.title']]
Categories.rename(columns={"id": "CategoryID", "snippet.title": "CategoryName"}, inplace = True)

Categories


# category=new_Item_id('id','snippet.title')
```

```python
rds_connection = connection_string = "postgres:Tr8ck321#@localhost:5432/Youtube2"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

```python
US_videos_cleaned.to_sql(name='Videos', con=engine, if_exists='append', index=False)
```

```python
Categories.to_sql(name='Categories', con=engine, if_exists='append', index=False)
```
